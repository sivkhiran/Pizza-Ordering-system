**SORTING**<br>
For sorting I have used the Quicksort 

```python
def quickSort(list_, key=None, reverse=False):
    if not key:
        def key(x): return x
    if reverse:
        def compare(x, y): return key(x) > key(y)
    else:
        def compare(x, y): return key(x) < key(y)
    n = len(list_)
    quickSortHlp(list_, 0, n-1, compare)
    return list_


def quickSortHlp(list_, first, last, compare):
    if first < last:
        splitpoint = partition(list_, first, last, compare)
        quickSortHlp(list_, first, splitpoint-1, compare)
        quickSortHlp(list_, splitpoint+1, last, compare)


def partition(list_, first, last, compare):
    p = list_[first]
    l, r = first+1, last
    while True:
        while l <= r and compare(list_[l], p):
            l += 1
        while r >= l and compare(p, list_[r]):
            r -= 1
        if r < l:
            break
        list_[l], list_[r] = list_[r], list_[l]
        l += 1
        r -= 1

    list_[first], list_[r] = list_[r], list_[first]
    return r  
```
